# <p style="background-color: #1F3A5F; font-family:calibri; color:white; font-size:140%; font-family:Verdana; text-align:center; border-radius:15px 50px;">1.| Load & Clean the Dataset</p>

```python
import pandas as pd

# Load data
data_path="/kaggle/input/datasets/vivek468/superstore-dataset-final/Sample - Superstore.csv"
df = pd.read_csv(data_path, encoding='latin1')

# Preview
print(df.head())

# Basic cleaning
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

# Convert dates
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Create time features
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Month Name'] = df['Order Date'].dt.strftime('%b')

# Profit margin
df['Profit Margin'] = df['Profit'] / df['Sales']

# Check data
print(df.info())
```

# <p style="background-color: #1F3A5F; font-family:calibri; color:white; font-size:140%; font-family:Verdana; text-align:center; border-radius:15px 50px;">2.| Core Business Analysis</p>


# <b><span style='color:#1F3A5F'>1.1 |</span><span style='color:#1F3A5F'> Revenue Trends Over Time</span></b>

```python
monthly_sales = df.groupby(['Year', 'Month Name'])['Sales'].sum().reset_index()
print(monthly_sales)
```

# <b><span style='color:#1F3A5F'>1.2 |</span><span style='color:#1F3A5F'> Top-Selling Products</span></b>

```python
monthly_sales = df.groupby(['Year', 'Month Name'])['Sales'].sum().reset_index()
print(monthly_sales)
```

# <b><span style='color:#1F3A5F'>1.3 |</span><span style='color:#1F3A5F'> Category Performance</span></b>

```python
category_sales = df.groupby('Category')[['Sales', 'Profit']].sum()
print(category_sales)
```

# <b><span style='color:#1F3A5F'>1.4 |</span><span style='color:#1F3A5F'> Regional Performance</span></b>

```python
region_sales = df.groupby('Region')[['Sales', 'Profit']].sum()
print(region_sales)
```

# <b><span style='color:#1F3A5F'>1.5 |</span><span style='color:#1F3A5F'> Profitability</span></b>

```python
profit_margin = df.groupby('Category')['Profit Margin'].mean()
print(profit_margin)
```
